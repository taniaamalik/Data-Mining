# Metode Penyelesaian
Penulis menggunakan beberapa metode penyelesaian agar dapat membandingkan metode terbaik yang dapat menyelesaikan permasalahan dengan lebih efisien. Metode yang digunakan antara lain :
- Single Linkage : Metode ini didasarkan pada jarak minimum. Dimulai dengan dua obyek yang dipisahkan dengan jarak paling pendek maka keduanya akan ditempatkan pada klaster pertama, dan seterusnya. Metode ini dikenal pula dengan nama pendekatan tetangga terdekat.
- Complete Linkage : Metode pautan lengkap (complete linkage) didasarkan pada jarak maksimum
- Average Linkage : Dasarnya adalah jarak rata-rata antar observasi. pengelompokan dimulai dari tengan atau pasangan observasi dengan jarak paling mendekati jarak rata-rata.
- Weighted Linkage :
- Metode Ward : Jarak antara dua cluster dalam metode ini berdasarkan total sum of square dua cluster pada masing-masing variabel.

Penelitian tersebut di lakukan oleh Rendy Handoyo dan teman-temannya pada tahun 2014. Pada penelitian itu dihasilkan sebuah kesimpulan bahwa metode Single Linkage memiliki performansi yang lebih baik dibandingkan dengan 
metode K-means. Metode pengujian yang digunakan untuk mengukur kualitas cluster adalah Silhouette Coefficient dan Purity. Nilai Silhouette Coefficient Single Linkage selalu lebih unggul dibandingkan dengan K-Means. Pertambahan jumlah dokumen membuat nilai Silhouette Coefficient single linkage semakin kecil sedangkan K-means terkadang menghasilkan nilai yang negatif. Untuk nilai Purity , Single Linkage selalu bernilai 1 sedangkan K-Means tidak pernah bernilai 1. Hasil pertambahan jumlah cluster dan jumlah dokumen memberikan pengaruh terhadap nilai Silhouette Coefficient dan Purity . Hal ini berarti single linkage selalu menghasilkan dokumen yang sama, sedangkan K-means masih bercampur dengan dokumen yang lain.

# Faktor Covid-19 Di Indonesia

```python
# Libraries
import pandas as pd
# %matplotlib notebook
from matplotlib import pyplot as plt
from scipy.cluster import hierarchy
import numpy as np
 
# Data set
# url = 'https://python-graph-gallery.com/wp-content/uploads/mtcars.csv'
url = 'D:\SMSTR 6\DATMIN\PROJEK AKHIR\dataset_normalized\dataset_normalized.csv'
df = pd.read_csv(url)
df = df.set_index('model')
del df.index.name
 
# Calculate the distance between each sample
Z = hierarchy.linkage(df, 'ward')
#    single
#    complete
#    average
#    weighted
#    ward
 
# Orientation of the dendrogram
plt.figure(figsize=(10, 7))
hierarchy.dendrogram(Z, orientation="right", labels=df.index)
plt.title("Faktor Covid Dendogram : Ward Lingkage")
plt.show()
# or
# hierarchy.dendrogram(Z, orientation="left", labels=df.index)
```

# Kasus Covid-19 Di Indonesia

```python
# Libraries
import pandas as pd
from matplotlib import pyplot as plt
from scipy.cluster import hierarchy
import numpy as np
 
# Data set
# url = 'https://python-graph-gallery.com/wp-content/uploads/mtcars.csv'
url = 'D:\SMSTR 6\DATMIN\PROJEK AKHIR\Kasus per provinsi\Kasus per provinsi.csv'
df = pd.read_csv(url)
df = df.set_index('Provinsi')
del df.index.name
 
# Calculate the distance between each sample
Z = hierarchy.linkage(df, 'weighted')
#    single
#    complete
#    average
#    weighted
#    ward
 
# Orientation of the dendrogram
plt.figure(figsize=(10, 7))
# plt.figure(figsize=(10, 7))
hierarchy.dendrogram(Z, orientation="right", labels=df.index)
plt.title("Kasus Covid Dendogram : Weighted Lingkage")
plt.show()
# or
# hierarchy.dendrogram(Z, orientation="left", labels=df.index)
```
